```python
def extract_substring(row):
    x = row['FullGenericName']
    y = row['GenericName']
    z = x.replace(y, '').strip()
    return z
```

```python
def separate_shape_and_dose(medication):
    words = medication.split()
    dose_index = next((i for i, word in enumerate(words) if any(char.isdigit() for char in word)), None)
    shape = ' '.join(words[:dose_index])
    dose = ' '.join(words[dose_index:])
    if dose==shape:
        dose=None
    return {'shape': shape, 'dose': dose}
```

```python
df_shape_dose = pd.DataFrame(df['Shape/Dose'])
```

```python
df[['Shape', 'Dose']] = df['Shape/Dose'].apply(lambda x: pd.Series(separate_shape_and_dose(x)))  
```
